```python
# Import Modules

# import pandas and numpy
import pandas as pd
import os
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Set some pandas options for controlling output
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)
pd.set_option('display.float_format', '{:.2f}'.format)

# inline graphics
%matplotlib inline
```

Now I will import the auto-mpg.data file and print the first few rows to gain an initial understanding of the data

```python
mpg = pd.read_table("auto-mpg.data",
                    sep=r'\s+',
                    header=None,
                    na_values='?')  
print(mpg.shape)
mpg.head()
```

Now I will add column names inline with the names document to provide a more accurate view of the metrics in each column

```python
columns = [
    'mpg',          
    'cylinders',    
    'displacement', 
    'horsepower',   
    'weight',       
    'acceleration', 
    'model_year',   
    'origin',       
    'car_name'      
]

mpg.columns = columns

print(mpg.shape)
mpg.head()
```

With the added column names we have a more descriptive picture of what each metric represents which will make our future visualization efforts easier

Now lets check the horsepower column for missing values and data type.

```python
print(f"Missing horsepower values: {mpg['horsepower'].isnull().sum()}")
```

The 6 values does match up with the six ? rows in the dataset. Performing this visual confirmation on large datasets would not be possible, but where possible a visual confirmation does add peace of mind.

Now I will map the rows to their country of origin using the following mapping:
1:USA, 2:Europe, 3:Asia

```python
mpg['origin'] = mpg['origin'].map({
    1: 'USA',
    2: 'Europe',
    3: 'Asia'
})

print(mpg['origin'].value_counts())
mpg.head()
```

Now I will create a barchart histogram that shows the distribution of cylinders

```python
from collections import Counter

# Count cylinders
cylinder_counts = Counter(mpg['cylinders'])

plt.bar(
    [x for x in cylinder_counts.keys()],  
    cylinder_counts.values(),             
    width=0.85,                            
    edgecolor=(0, 0, 0)                   
)

plt.xticks([3, 4, 5, 6, 8])              
plt.xlabel("Number of Cylinders")
plt.ylabel("Number of Cars")
plt.title("Distribution of Cylinders")
sns.despine()
plt.tight_layout()
plt.show()
```

Now I will create a scatterplot to show the relationship between horsepower and weight

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=mpg,
    x='horsepower',
    y='weight',
    alpha=0.6
)
sns.despine()
plt.tight_layout()
plt.title("Horsepower vs Weight")
plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.show()
```

Now I will use matplotlib and seaborn to visualize the relationship between the number of cylinders a car engine has and its affect on MPG using a boxplot and see what conclusions I can draw from this visual.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=mpg,
    x='cylinders',
    y='mpg',
    palette='viridis'
)
sns.despine()
plt.tight_layout()
plt.title("MPG Distribution by Number of Cylinders")
plt.xlabel("Number of Cylinders")
plt.ylabel("Miles Per Gallon")
plt.show()
```

This boxplot shows an inverse relationship between the number of cylinders a car has and its miles per gallon. 
- 3-cylinder engines show a consistent but modest fuel economy around 20 MPG. 
- 4-cylinder engines achieve the highest median MPG ~28, with the widest efficiency. 
- 5- and 6-cylinder engines fall between 4 and 8 cylinders as expected, with them sitting between the two literally. 
- 8-cylinder engines have the lowest median MPG ~15, confirming that 8-cylinder engines are the least fuel efficient; they also show the most outliers above 25 MPG.
