```python
# Import Modules

# import pandas and numpy
import pandas as pd
import os
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Set some pandas options for controlling output
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)

# inline graphics
%matplotlib inline
pd.set_option('display.float_format', '{:.2f}'.format)
```

Now I will import the data file and print the head to gather initial information about this data set

```python
mp = "auto-mpg.data"

if not os.path.exists(mp):
    url = "https://raw.githubusercontent.com/dwaynewebb86/IS362_Week8/main/auto-mpg.data"
    mpg = pd.read_csv(url, 
                      sep=r'\s+', 
                      header=None, 
                      na_values='?')
    mpg.to_csv(mp, index=False)
    print("Downloaded and saved locally")
else:
    mpg = pd.read_csv(mp)
    print("Loaded from local file")

print(mpg.shape)
mpg.head()
```

Now I will name the columns to give us a better understanding of what each metric represents

```python
mpg.columns = [
    'mpg', 'cylinders', 'displacement',
    'horsepower', 'weight', 'acceleration',
    'model_year', 'origin', 'car_name'
]

print(mpg.columns.tolist())
mpg.head()
```
